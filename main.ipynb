```python
#imports
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, BasicAer, transpile
from qiskit.quantum_info import Statevector
from modules.classic.decoder import getBits
from modules.quantum.bell import create_bell
from modules.quantum.encode import quantum_encode
from modules.quantum.decode import quantum_decode
from modules.quantum.VectorSimul import vector_simulator
```

```python
image = 'assets/image-for-quantum.png'
```

```python
#Define Backend
backend = BasicAer.get_backend("statevector_simulator")
```

```python
#define Quantum & Classical registers
Qbits = QuantumRegister(2)
Cbits = ClassicalRegister(2)
```

```python
bitPairs = getBits(image)
```

```python
print(bitPairs)
```

```python
#define quantum circuit
qc = create_bell(QuantumCircuit,Qbits,Cbits)
```

```python
qList = [] #save data after all quantum decode and encoding
```

```python
pair = bitPairs[0]
print(pair)
qc = quantum_encode(qc,1,pair)
qc = quantum_decode(qc)
```

```python
qc.measure(1,1)
```

```python
measured = vector_simulator(qc,backend,transpile,Statevector)
```

```python
if measured[0].real == 1:
    qList.append(0)
else:
    qList.append(1)
```

```python
print(qList)
```
